```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
covid = pd.read_excel("owid-covid-data.xlsx")

covid = covid[['location','date','total_cases','new_cases','total_deaths','new_deaths']]
covid.head()
```

```python
covid = covid[covid['location'] == 'Afghanistan']

covid.plot(x = "date",y="new_deaths", figsize = (12,12))
plt.show()
```
